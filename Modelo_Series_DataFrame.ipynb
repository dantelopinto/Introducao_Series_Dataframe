Abaixo criamos uma Series notas, o index desta Series √© a coluna √† esquerda, que vai de 0 a 4 neste caso, que o pandas criou automaticamente, j√° que n√£o especificamos uma lista de r√≥tulos.</font>

### <font color='black'>**Dataframe**: Um **dataframe** √© semelhante √† pasta de trabalho do Excel - voc√™ tem nomes de colunas e linhas, que podem ser acessadas com o uso de n√∫meros de linhas. A diferen√ßa essencial √© que os nomes de colunas e os n√∫meros de linha s√£o conhecidos como √≠ndice de coluna e linha, no caso de dataframe.</font>

### <font color='black'>**S√©ries** e **Dataframe** formam o modelo de dados principal do Pandas em Python. Os conjuntos de dados s√£o os primeiros a ler esses dataframes e, em seguida, v√°rias opera√ß√µes (por exemplo, agrupar por, agrega√ß√£o etc.) podem ser aplicadas com muita facilidade √†s suas colunas.</font>

```python
stock_dictionary = {'AMZN': pd.Series([346.15,0.59,459,0.52,589.8,158.88],
                           index = ['Closing price','EPS','Shares Outstanding(M)',
                                  'Beta', 'P/E','Market Cap(B)']),
         'GOOG': pd.Series([1133.43,36.05,335.83,0.87,31.44,380.64],
                            index = ['Closing price','EPS','Shares Outstanding(M)',
                                  'Beta','P/E','Market Cap(B)']),
         'FB': pd.Series([61.48,0.59,2450,104.93,150.92],
                            index = ['Closing price','EPS','Shares Outstanding(M)',
                                  'P/E', 'Market Cap(B)']),
         'YHOO': pd.Series([34.90,1.27,1010,27.48,0.66,35.36],
                            index=['Closing price','EPS','Shares Outstanding(M)',
                                  'P/E','Beta', 'Market Cap(B)']),
         'TWTR':pd.Series([65.25,-0.3,555.2,36.23],
                            index=['Closing price','EPS','Shares Outstanding(M)',
                                  'Market Cap(B)']),
         'AAPL':pd.Series([501.53,40.32,892.45,12.44,447.59,0.84],
                            index=['Closing price','EPS','Shares Outstanding(M)','P/E',
                                  'Market Cap(B)','Beta'])}
df = pd.DataFrame(stock_dictionary)
df
```

```python
df = pd.DataFrame(stock_dictionary, index = ['Closing price',
                                 'EPS',
                                 'Shares Outstanding(M)',
                                 'P/E',
                                 'Market Cap(B)',
                                 'Beta'],
                        columns = ['FB','TWTR','SCNW'])
df
```

## <font color='grey'>Group By e Reshape dos Dados</font>

```python
cidadesDigitais = pd.read_csv('/content/cidades_digitais.csv', sep = ';', encoding = 'latin-1')
```

```python
cidadesDigitais.head()
```

```python
cidadesDigitais.count()
```

```python
cidadesUF_1 = cidadesDigitais.groupby('UF')
```

```python
type(cidadesUF_1)
```

```python
cidadesUF_1.groups
```

```python
len(cidadesUF_1.groups)
```

```python
ordenaUF = cidadesUF_1.size()
```

```python
ordenaUF
```

```python
ordenaUF.sort_values(ascending = False)
```

```python
cidadesUF_2 = cidadesDigitais.groupby(['UF','STATUS'])
```

```python
total = cidadesUF_2.size()
total.sort_values(ascending = False)
total
```

```python
cidadesUF_3 = cidadesDigitais.set_index('STATUS')
cidadesUF_3 = cidadesUF_3.groupby(level = 'STATUS')
cidadesUF_3.head()
```

```python
cidadesUF_4 = cidadesDigitais.set_index(['STATUS', 'UF'])
cidadesUF_4
```

```python
cidadesUF_4 = cidadesDigitais.set_index(['UF', 'STATUS'])
cidadesUF_4 = cidadesUF_4.groupby(level = ['UF', 'STATUS'])
print(cidadesUF_4)
```

```python
cidadesUF_4.sum()
```

```python
cidadesUF_4.aggregate(np.sum)
```

```python
cidadesUF_4.aggregate([np.sum, np.mean, np.size])
```

```python
cidadesUF_5 = cidadesDigitais.set_index([ 'STATUS', 'CIDADE', 'VALOR_TOTAL_PREVISTO', 'VALOR_INVESTIDO', 'POPULA√á√ÉO', 'PONTOS_ATENDIDOS'])
```
